# Gráficos

Aqui procurarei apresentar alguns modelos de gráficos que podem ser úteis para uso nas análises de machine learning.

```julia
using Plots, StatsPlots, Queryverse, KernelDensity, Distributions
```

```julia
dados = load("iris.csv") |> DataFrame;
first(dados,3)
```

## Heatmap

pode ser muito útil para análise de correlação

heatmap(x, y, data)
  - x -> nome das classe no eixo x
  - y -> nome das classe no eixo y
  - data -> matriz de dados que será plotada

caso seja de interesse girar os nomes do eixo x pode ser utilizado o atributo __xrotation=__ e inserido o valor de rotação desejado.

```julia
input_cor = abs.( cor(dados[!, 1:end-1] |> Matrix) );
```

```julia
heatmap(String.(names(dados)[1:end-1]), String.(names(dados)[1:end-1]),
        input_cor, xrotation=60,  c=ColorGradient([:white,:red,:blue]),
        xticks = :all, yticks = :all)
```

## Boxplot/Violin/dotplot

Pode ser útil na busca por outliers em um conjunto de dados, ou para buscar tendências entre variáveis de classe na entrada e uma saída numérica.

```julia
@df dados boxplot(:variety, :sepallength, group = :variety, legend = :topleft)
```

```julia
@df dados violin(:variety, :sepallength, group = :variety, legend = :topleft)
```

```julia
@df dados dotplot(:variety, :sepallength, group = :variety, legend = :topleft)
```

# Distribuições

Podemos plotar o histograma dos dados facilmente

```julia
@df dados histogram(:sepallength)
```

É possível ainda faze algo similiar a função distplot de um pacote do Python, para isso vamos fazer uso agora do pacote KernelDensity e estimar com a função kde, após isso faremos um histograma e um plot normal, sendo que nosso histograma será normalizado.

```julia
@df dados histogram(:sepallength, normalize=true)
@df dados plot!( kde(:sepallength), linewidth = 2, color=:red)
```

```julia
x = rand(Normal(), 100)
@df dados qqplot(Normal, x )
```

```julia
@df dados scatter(:petallength,:sepallength, group = :variety, legend = :topleft)
```
